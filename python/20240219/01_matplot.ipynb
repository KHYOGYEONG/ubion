라인 그래프
- plot({1차원 데이터 형태})
- plot 1차원 데이터 인자의 개수가 1개
- y축 데이터로 대입
- x축은 0부터 1씩 증가하는 데이터 생성
- 인자의 개수가 2개인 경우
- 첫번째 인자가 x축 데이터
- 두번째 인자가 y축 데이터

```python
# 1차원 데이터의 경우
data = [5, 3, 7, 1]

plt.plot(data)
plt.show()
```

```python
## 2개의 그래프를 출력
plt.plot(data)
plt.plot([10,2,4,7])
plt.show()
```

```python
# x축 데이터 넣기
plt.plot([1, 2, 3, 4], data)
plt.show()
```

```python
# 라인 그래프에서 라인 색 변경하기
plt.plot(data, "#FF0066")
plt.show()
```

```python
# 선의 종류 변경
plt.plot(data, "--")
plt.show()
```

```python
# x, y에 라벨 설정
# 그래프의 제목 설정
plt.plot(data)
plt.xlabel('X')
plt.ylabel('Y')
plt.title("Graph")
plt.show()
```

```python
# 막대 그래프
# bar({x축 데이터},{y축 데이터})
x = [1, 2, 3, 4]
y = [80, 90, 70, 80]

plt.bar(x, y) # x와 y값 둘 다 넣어야댐. 하나만 넣으면 에러발생
plt.show()
```

```python
# 막대의 색상
plt.bar(x, y, color=['red', 'y'])
plt.show()
```

```python
# 막대의 너비 변경
plt.bar(x, y, width=0.4)
plt.show()
```

```python
# 한글이 꺠짐 # 아래 과정을 통해 한글 나오게 설정해줘야 함
plt.bar(['가', '나', '다', '라'], y)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
```

```python
## csv 폴더 안에 있는 drinks.csv 로드
drinks = pd.read_csv("../../../csv/drinks.csv")
```

```python
drinks.info()
```

```python
## 데이터에서 상관계수를 출력
## 데이터프레임 안에 내장된 corr() 함수 사용
corr = drinks.iloc[:, 1:5].corr(method='pearson')
corr
```

```python
## 히트맵 출력
cols_view = ['beer', 'spirit', 'wine', 'alcohol']

hm = sns.heatmap(
    corr.values,
    cbar = True, # 사이드바 유무
    annot = True, # 히트맵에 숫자 표시 유무
    square = True, # 히트맵을 정사각형 표시 유무
    fmt = ' .3f', # 숫자의 소수점 자리수
    annot_kws = {'size' : 20}, # 숫자의 크기
    xticklabels = cols_view,
    yticklabels = cols_view
    )
plt.tight_layout()
plt.show()
```

```python
## 데이터의 분포도를 시각화
sns.pairplot(
    drinks.loc[:,'beer_servings' : 'total_litres_of_pure_alcohol'],
    height = 2.5,
    corner=True,
    kind = 'reg',
    plot_kws={ # 직선을 넣어줘
        'line_kws' : {
            'color' : 'red'
        }
    }
)

plt.show()
```
